## World Cup Problem Revisisted

In Chapter 8, we solved the World Cup Problem using a Poisson process to model goals in a soccer game as random events that are equally likely to occur at any time. We used a gamma distribution to represent the prior distribution of $\lambda$, the goal-scoring rate. And we used a Poisson distribution to compute the probability of $k$, the number of goals scored.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import gamma

alpha = 1.4
dist = gamma(alpha)

# grid approximation:
from utils import pmf_from_dist

lams = np.linspace(0, 10, 101)
prior = pmf_from_dist(dist, lams)

from scipy.stats import poisson

k = 4
likelihood = poisson(lams).pmf(k)

# update
posterior = likelihood * prior
posterior.normalize()
```

So far this should be very familiar! Now let's solve it using conjugate priors

## Conjugate Priors
One reason that we used a gamma distribution for the prior is that it is the "conjugate prior" of the Poisson distribution. Let's see what that means!

To do so, we need to move to the 2-parameter version of the gamma distribution. Shockingly, the two parameters are $\alpha$ and $\beta$.

```python
def make_gamma_dist(alpha, beta):
    dist = gamma(alpha, scale=1/beta)
    dist.alpha = alpha
    dist.beta = beta
    return dist

# Here's the prior distribution with `alpha=1.4` again and `beta=1`
alpha = 1.4
beta = 1

prior_gamma = make_gamma_dist(alpha, beta)
prior_gamma.mean()
```

Now: the claim is that doing the update with `k` goals just by making a gamma distribution with parameters `alpha + k` and `beta + t`.

```python
def update_gamma(prior, data):
    """ Update a gamma distribution """
    k, t = data
    alpha = prior.alpha + k
    beta = prior.beta + t
    return make_gamma_dist(alpha, beta)

# here's how we update it with $k=4$ goals in $t=1$ games
data = 4, 1
posterior_gamma = update_gamma(prior_gamma, data)

# let's confirm that this works
posterior_conjugate = pmf_from_dist(posterior_gamma, lams)
```

```python
posterior.plot()
posterior_conjugate.plot()
```

My goodness! They are the same!

# What the Actual?

Let's take a look at what is going on:

PDF of the gamma prior:
$$
\lambda^{\alpha - 1}e^{-\lambda \beta}
$$

Now say the team scores $k$ goals in $t$ games. The probability of this data is given by the PMF of the Poisson distribution, which is a function of $k$ with $\lambda$ and $t$ as parameters:

$$
\lambda^{k}e^{-\lambda t}
$$

So their product is

$$
\begin{align*}
\lambda^{k}e^{-\lambda t} \ast \lambda^{\alpha - 1} e^{-\lambda \beta} &= \\
&=  \lambda^{\alpha - 1 + k} e^{-\lambda(\beta + t)}
\end{align*}
$$

Which is an unnormalized gamma distribution with parameters $\alpha + k$ and $\beta + t$.

This formulation also gives us some insights into what the parameters mean: $\alpha$ is the related to the number of successes, and $\beta$ is the elapsed time.

## Binomial Likelihood
Let's look at the Euro problem again! When we solved it with a grid method, we started with a uniform prior:

```python
from utils import make_uniform

xs = np.linspace(0, 1, 101)
uniform = make_uniform(xs, 'uniform')

# now compute the likelihood, which was 140 heads out of 250 attempts
from scipy.stats import binom

k, n = 140, 250
xs = uniform.qs
likelihood = binom.pmf(k, n, xs)

# now compute the posterior
posterior = uniform * likelihood
posterior.normalize()
```

Again, we can solve this more efficiently by using a conjugate prior. The conjugate prior of the binomial distribution is the beta distribution.

The beta distribution is bounded between 0 and 1 and takes two parameters, $\alpha$ and $\beta$.

```python
from scipy.stats import beta
import scipy

def make_beta(alpha, beta):
    """ Makes a beta object """
    dist = scipy.stats.beta(alpha, beta)
    dist.alpha = alpha
    dist.beta = beta
    return dist

# the uniform distribution can be represented as a beta(1, 1)
alpha = 1
beta = 1
prior_beta = make_beta(alpha, beta)
```

Again, let's take a look at how we can see that they are conjugate

The PDF of the beta is given by
$$
x^{\alpha - 1}(1-x)^{\beta - 1}
$$
(again, without the normalizing parameter)

The pmf of the binomial distribtion with $k$ successes out of $n$ attempts and a probability of $x$:

$$
x^k (1-x)^{n - k}
$$

Their product is
$$
x^{\alpha - 1 + x} (1 - x)^{beta - 1 + n - k}
$$

which is an unnormalized beta distribution with parameter $\alpha + k$ and $\beta + n - k$.

So, if we observe $k$ successes in $n$ trials, we can do the update by making a beta distribution with parameters `alpha + k` and `beta + n - k`.

```python
def update_beta(prior, data):
    k, n = data
    alpha = prior.alpha + k
    beta = prior.beta + n - k
    return make_beta(alpha, beta)
```

Again, this parameterization gives us insights into the parameters: $\alpha$ is the number of successes, and $\beta$ is the number of failures.

```python
data = 250, 140
posterior_beta = update_beta(prior_beta, data)

# to confirm
posterior_conjugate = pmf_from_dist(posterior_beta, xs)

posterior_conjugate.plot()
posterior.plot()
```

## Lions and Tigers and Bears

Suppose that we visit a wild animal preserve where we know that the only animals are lions, tigers, and bears, but we don't know how many of each there are. During the tour we see three lions, two tigers, and one bear. Assuming that every animal had an equal change to appear in our sample, what is the probability that the next anumals that we seee is a bear?

To answer this question, we'll use the data to estimate the prevalence of each species; that is, what fraction of the animals belong to each species. If we know the prevalence, we can use the multinomial distribution to compute the probability of the data.

For example, say that we know that the fraction of lions, tigers, and bears is 0.4, 0.3, and 0.3 respectively.

```python
from scipy.stats import multinomial

ps = 0.4, 0.3, 0.3
data = 3, 2, 1
n = np.sum(data)

multinomial.pmf(data, n, ps)
```

Now, we could choose a prior for the prevalences and do a Bayesian update using the multinomial distribution to compute the distribution of the data. However, there's an easier way - the multinomial distribution has a conjugate prior, the Dirichlet distribution.

## The Dirichlet Distribution

The Dirichlet distribution is a multivariate distribution, like the multivariate normal we used with the penguins.

In that case, the quantities in the distribution were pairs of flipper lengths and culmen lengths, and the parameters of the distribution were a matrix of variances and a vector of means.

In a Dirichlet distribution, the quantities are vectors of probabilities, $\vec{x}$, and the parameter is a vector $\vec{\alpha}$.

```python
# let's see an example!

from scipy.stats import dirichlet

alpha = 1, 2, 3
dist = dirichlet(alpha)

# since we provided three parameters, the result is a distribution of three variables. Suppose we draw a random value from this distribtion:
dist.rvs()
```

The result is an array of three valyes. They are bounded in $[0, 1]$ and they always add up to 1. Thus, they can be interpreted as the probabilities of a set of outcomes that are mutually exclusive and collectively exhaustive.

Let's see what the distributions of these values look like:

```python
sample = dist.rvs(1000)

from empiricaldist import Cdf
cdfs = [Cdf.from_seq(col) for col in sample.transpose()]

for cdf in cdfs:
    cdf.plot()
```

As it turns out, these distributions are beta! The following takes a sequence of parameters, $alpha$, and computes the marginal distribution of variable $i$.

```python
from empiricaldist import Pmf

def marginal_beta(alpha, i):
    """ Computes the ith marginal of a Dirichlet distribution """
    total = np.sum(alpha)
    return make_beta(alpha[i], total - alpha[i])

# for the three variables that we have:
marginals = [ marginal_beta(alpha, i) for i in range(len(alpha))]

for marginal in marginals:
    pmf_from_dist(marginal, np.linspace(0, 1, 101)).make_cdf().plot()
for cdf in cdfs:
    cdf.plot()
```

So again, we see that they are basically the same. This is some evidence that the marginals of the Dirichlet distribution are beta. That is very useful because the Dirichlet distribution is the prior of the multinomial distribution.

If the prior distribution is Dirichlet with parameter vector `alpha` and the data is a vector of observations `data`, the posterior distribution is Dirichlet with parameter vector `alpha + data`.

### Exercise 18-3

Problem from 3Blue1Brown:

> You are buying a product online and you see three sellers offering the same product at that same price. One of them has a 100% positive ratings, but with only 10 reviews. Another has a 96% positive rating, with 50 reviews. Yet another has a 93% positive rating, but with 200 total reviews. Which one should you buy from?

Let's think about how to model this scenario. Suppose that each seller has some unbknown probability $x$ of providing satisfactory service and getting a positive rating, and we want to choose the seller with the highest value of $x$. 

This is not the only model for this scenario, and it is not even necessarily the best. An alternative would be something like item response theory, where sellers have variying ability to provide satisfactory service and customers have varying difficulty of being satisfied.

But the first model has the virtue of simplicity, so let's stick with it.

1. As a prior, use a beta distribution with $\alpha=8$ and $\beta=2$.
1. Use the data to update the prior for each of the three sellers and plot the posterior distribution. Which seller has the highest posterior mean?
1. How confident should we be about our choice? That is, what is the probability that the seller with the highest posterior mean actually has the highest value of $x$?
1. Consider a beta prior with `alpha=0.7` and `beta=0.5`. What does this prior look like and what does it imply about the sellers?
1. Run the analysis again with this prior and see what effect it has on the results.

### Solution 18-3

```python
class Seller:
    def __init__(self, alpha, beta):
        self.alpha = alpha
        self.beta = beta
        
prior_alpha = 8
prior_beta = 2

def make_seller_prediction(prior_alpha, prior_beta):
    print(f"Making prediction for prior alpha {prior_alpha} and prior beta {prior_beta})")
    for rating, number_ratings in [
        [1, 10],
        [0.96, 50],
        [0.93, 200]
    ]:
        seller_alpha = rating * number_ratings
        seller_beta = (1 - rating) * number_ratings
        seller = Seller(seller_alpha, seller_beta)
        posterior_beta = beta(seller.alpha + prior_alpha, seller.beta + prior_beta)
        basis = np.linspace(0, 1, 101)
        raw_pmf = np.nan_to_num(posterior_beta.pdf(basis), posinf=0)
        normalized_pmf = raw_pmf / raw_pmf.sum()
        pmf = Pmf(normalized_pmf, basis)
        fig, ax = plt.subplots()
        ax.set_title(f"Seller: {rating * 100}% Positive With {number_ratings} Ratings")
        ax.set_xlabel("Probability of Positive Experience")
        ax.set_ylabel("Probability Density")
        ax.plot(pmf.qs, pmf.ps)
        print(f"Data for seller with {rating * 100}% positive with {number_ratings} ratings")
        print(f"Mean: {pmf.mean() * 100:.2f}%")
        print(f"95% CI: {pmf.credible_interval(0.95)}")
        print("-------------")
```

```python
make_seller_prediction(8, 2)
```

```python
make_seller_prediction(0.7, 0.5)
```
