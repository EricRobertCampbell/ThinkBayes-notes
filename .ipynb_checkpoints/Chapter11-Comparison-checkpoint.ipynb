Joint distributions, which are somehow involved with distributions of more than one variable.

## Outer Operations

Many operations can be expressed as the "outer product" of some sequences.

For instance (apparently) the outer product of two sequences `[1, 3, 5]` and `[2, 4]` is the array that contains the product of each of the constituent elements.

```python
import numpy as np

x = [1, 3, 5]
y = [2, 4]

X, Y = np.meshgrid(x, y)
print("X", X)
print("Y", Y)
```

`X` contains copies of `x`, arranged in as many rows as there are entries in `y`

`Y` contains copies of `y`, arranged in as many columns as there are entries in `x`

Because the new arrays (`X` and `Y`) are the same size, we can use them as operands for arithmetic functions like multiplication!

```python
X * Y
```

That is the outer product of `x` and `y`. We can see this more clearly if we use a dataframe:

```python
import pandas as pd

df = pd.DataFrame(X * Y, columns=x, index=y)
df
```

## How tall is A?

Suppose I choose two perople from the populate of adult males in the US; say A and B. If we see that A is taller than B, how tall is A?

To answer the question:
1. We'll use background information about the height of adult males in the US to for a prior distribution of height
1. We'll construct a joint prior distribution of height for A and B
1. Update the prior with the information that A is taller
1. Extract the posterior distribution of height for A

In the US the height of adult males is pretty much normal, with $\mu = 178cm$ and $\sigma = 7.7cm$. 

```python
mean = 178
qs = np.arange(mean - 24, mean + 24, 0.5) # plus and minus about 3 standard deviations

from scipy.stats import norm
std = 7.7
ps = norm(mean, std).pdf(qs)

from empiricaldist import Pmf

prior = Pmf(ps, qs)
prior.normalize()
prior.plot()
```

## Joint Distribution

The next step is to construct the distribution that represents the probabilty of every pair of heights, which is called the joint distribution.

The elements of the joint distribution are 
$$
P(A_x\ \text{and}\ B_y)
$$

which is the probability that A has height $x$ and B has height $y$.

Since the height of A and B are independent, 
$$
P(A_x\ \text{and}\ B_y) = P(A_x)P(B_y)
$$

We can compute the joing probability by computing the outer product of each distribution!

```python
def make_joint(pmf1, pmf2):
    """ Make the joint distribution by computing the outer product """
    X, Y = np.meshgrid(pmf1, pmf2)
    return pd.DataFrame(X * Y, columns=pmf1.qs, index=pmf2.qs)

# Luckily for us, the pmfs are the same
joint = make_joint(prior, prior)
joint.shape
```

So this is a 96 x 96 matrix

If the priors are normalized, then the joint probability is also normalized:

```python
joint.to_numpy().sum()
```

## Visualizing the Joint Distribution

```python
import matplotlib.pyplot as plt

def joint_plot(joint, cmap="Blues"):
    """ Plot the joint distribution with a colour mesh """
    vmax = joint.to_numpy().max() * 1.1
    plt.pcolormesh(joint.columns, joint.index, joint, cmap=cmap, vmax=vmax, shading='nearest')
    plt.colorbar()
    
joint_plot(joint)
```

```python
# We could also use a contour plot
def plot_contour(joint):
    """ Plot a joint distribtution using a contour plot """
    plt.contour(joint.columns, joint.index, joint, linewidths=2)
    
plot_contour(joint)
```

## Likelihood

Now that we have the joint distribution, we can update it with the date (A is taller than B).

```python
x = joint.columns
y = joint.index

X, Y = np.meshgrid(x, y)

# X - possible values for A
# Y - possible values for B

A_taller = (X > Y)

# an array of booleans
A_taller.dtype
```

To compute the likelihoods, use np.where to make an array with value 1 when A is taller and 0 otherwise

```python
a = np.where(A_taller, 1, 0)
likelihood = pd.DataFrame(a, index=x, columns=y)
joint_plot(likelihood)
```

## The Update

Now we have the prior, we have the likelihood - let's do the update!

```python
posterior = joint * likelihood

def normalize(joint):
    """ Normalize a joint distribution """
    total_weight = joint.to_numpy().sum()
    joint /= total_weight
    return joint

normalize(posterior)
```

```python
joint_plot(posterior)
```

This basically looks like out original, except the values where A is not taller have been eliminated (and then also renormalized)

To find the probabilities for A's height, we will iterate over all of the heights of B and find the probability of A having some height conditional on all of those heights (we'll integrate them out):

```python
column_sums = posterior.sum(axis=0)
column_sums.head()
```

This is a Series that contains the values of the different heights and their probabilities - in other words, a distribution for A!

```python
marginal_A = Pmf(column_sums)
marginal_A.plot()
```

```python
# We can do the same for B
row_sums = posterior.sum(axis=1)
marginal_B = Pmf(row_sums)
marginal_B.plot()
```

```python
def marginal(joint, axis):
    """
    Given a joint distribution, find the marginal distribution of one of the axes - 
    0 for the first one (x), and  
    1 for the second one (y)
    """
    return Pmf(joint.sum(axis=axis))
```

```python
print('Original:', prior.mean())
print('Marginal A:', marginal_A.mean())
print('Marginal B:', marginal_B.mean())
```

```python
marginal_A.plot()
marginal_B.plot()
prior.plot()
```

```python
print("Marginal A std:", marginal_A.std())
print("Marginal B std:", marginal_B.std())
```

So we can see that the std as smaller - we are more certain about their heights now.

## Conditional Posteriors

Now suppose that we measure A and find him to be 170cm tall.

```python
column_170 = posterior[170] # column of B's heights and probabilities when A = 170
column_170
```

```python
cond_B = Pmf(column_170)
cond_B.normalize()
```

```python
prior.plot()
marginal_B.plot()
cond_B.plot()
```

### Exercise 11-2
> Suppose we have established that A is taller than B, but we don't know how tall B is. Now we choose a random woman, C, and find that she is shorter than A by at least 15cm. Compute the posterior distributions for the heights of A and C.
>
> The average height for women in the US is 163cm; the standard deviation is 7.3 cm.

```python
### Exercise 11-2 Solutio

# Create the normal distribution for women's heights
women_mean = 163
women_std = 7.3
qs = np.arange(women_mean - 24, women_mean + 24, 0.5)
ps = norm(women_mean, women_std).pdf(qs)
c_prior = Pmf(ps, qs)
c_prior.plot()
```

```python
# Now we make the joint distribution for A and C
# Of course, we remember that we actually have information about A from B
joint_a_c = make_joint(marginal_A, c_prior)
joint_plot(joint_a_c)
joint_a_c.shape
```

```python
# Now we need the likelihood - that is all of the cases where A - C > 15
x = joint_a_c.columns
y = joint_a_c.index

X, Y = np.meshgrid(x, y)
a = np.where(X - Y >= 15, 1, 0)
new_likelihood = pd.DataFrame(a, index=y, columns=x)
joint_plot(new_likelihood)
```

```python
# Now the update
posterior_a_c = joint_a_c * new_likelihood
normalize(posterior_a_c)
```

```python
joint_plot(posterior_a_c)
```

```python
marginal_A = marginal(posterior_a_c, 0)
marginal_C = marginal(posterior_a_c, 1)
marginal_A.plot()
marginal_C.plot()
print(f"Marginal A mean: {marginal_A.mean()}")
print(f"Marginal C mean: {marginal_C.mean()}")
```

### Exercise 11-3
> The Elo rating system is a way to quantify the skill level of players for a game like chess. It is based on a model of the relationship between the ratings of players and the outcome of a game. Specifically, for the skill level $R_A$ of player A and $R_B$ of player B, the probability of A beating B is
$$
P(\text{A beats B}) = \frac{1}{1+10^{\frac{R_B - R_A}{400}}}
$$
>
>The parameters of 10 and 400 are arbitrary; they are chosen to determine the range of the ratings. In chess, the range is from 100 to 2800.
>
>Suppose that A has a current rating of 1600, but we are not sure that it is accurate. We could describe their true rating with a normal distribution with mean of 1600 and a standard deviation of 100, to indicate our uncertainty.
>
>B has a current rating of 1800, with the same level of uncertainty.
>
>Then A and B play and A wins. How should we update their ratings?

```python
def calc_prob(ra, rb):
   return 1 / (1 + 10 ** ((rb - ra) / 400))
calc_prob(100, 0)
```

```python
mean_a = 1600
std_a = 100
qs = np.arange(1300, 1900, 10)
ps = norm(mean_a, std_a).pdf(qs)
prior_a = Pmf(ps, qs)
prior_a.normalize()
prior_a.plot()

mean_b = 1800
std_b = 100
qs = np.arange(1500, 2100, 10)
ps = norm(mean_b, std_b).pdf(qs)
prior_b = Pmf(ps, qs)
prior_b.normalize()
prior_b.plot()
```

```python
# Now we make the joint distribution for A and B (for each pair of hypotheses about the rating)

joint = make_joint(prior_a, prior_b)
joint_plot(joint)
joint.shape
```

```python
x = joint.columns
y = joint.index
X, Y = np.meshgrid(x, y)
diff = X - Y
a = 1 / (1 + 10 ** (-diff / 400))
likelihood = pd.DataFrame(a, columns=x, index=y)
joint_plot(likelihood)
```

```python
posterior = joint * likelihood
normalize(posterior)
joint_plot(posterior)
```

```python
marginal_A = marginal(posterior, axis=0)
marginal_B = marginal(posterior, axis=1)
print(f"Marginal A mean: {marginal_A.mean()}")
print(f"Marginal B mean: {marginal_B.mean()}")
marginal_A.plot()
marginal_B.plot()
```
