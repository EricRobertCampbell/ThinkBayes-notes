## Log Odds

Author went to a class on Theory of Computation. Three other students arrive, all women. Given that the computer science program was about 83\% male, he began to believe that he was in the wrong class.

Let's use this to demonstrate Bayes' rule on a logarithmic scale and show how it relates to logistic regression.

`H` - the hypothesis that he was in the right room, and `F` to represent the overservation that the first other student was female, we can write Bayes' Rule like this:

$$
O(H|F) = O(H)\frac{P(F|H)}{P(F|\text{not } H)}
$$

Before he saw the other student, he was confident that he was in the right class, so

$$
O(H) = 10:1 = 10
$$

If I was in the right room, the likelihood of the first female student was about 17\%. If I was not in the right room, the likelihood of the first female student was more like 50\%.

$$
\frac{P(F|H)}{P(F|\text{not }H)} = \frac{17}{50} \approx \frac{1}{3}
$$

Applying Bayes' Rule:

$$
O(H|F) = \frac{10}{3}
$$

After two students:

$$
O(H|FF) = \frac{10}{9}
$$

after 3:

$$
O(H|FFF) = \frac{10}{27}
$$

at that point, he was right to suspect that he was in the wrong classroom.

The following table shows the odds after each update, the probability, and the change in probability, expressed in percentage points:

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(columns=['odds', 'prob', "prob_diff"])
df.loc['prior'] = [10, 1 / (10 + 1), 0]

for num_students in range(1, 4):
    prev_data = df.iloc[num_students - 1]
    new_odds = prev_data['odds'] * (1/3)
    new_prob = 1 / (1 + new_odds)
    df.loc[f"{num_students} students"] = [new_odds, new_prob, prev_data['prob'] - new_prob]
df
```

Now let's add the log odds to the table:

```python
df['log odds'] = np.log(df['odds'])
df['log odds diff'] = df['log odds'].diff()
df.fillna('--')
```

So we can see that the difference in the logg odds is a constance: $-1.098 = \log (1/3)$. In fact, we can see that this will always be the case by taking the log of Bayes' Rule:

$$
\log O(H|F) = \log O(H) + \log \left(\frac{P(F|H)}{P(F|\text{not }H} \right)
$$

On a log odds scale, a Bayesian update is additive! So if $F^x$ means the $x$ female students arrive while I'm waiting, then posterior logs odds that I'm in the right room are

$$
\log O(H|F^x) = \log O(H) + x\log \left( \frac{P(F|H)}{P(F| \text{not }H)} \right)
$$

In this example the relationship is exact, but even when it is not quite, it is common to use a linear function to model the relationship between an explanatory variable, $x$, and a dependent variable expressed in log odds like this:

$$
\log O(H|x) = \beta_0 + \beta_1 x
$$

Where $\beta_0$ and $\beta_1$ are unknown parameters:
- The intercept, $\beta_0$, is the log odds of the hypothesis when $x=0$
- The slope, $\beta_1$m is the log of the likelihood ratio

## The Space Shuttle Problem

Let's solve an example of logistic regression from Cameron Davidson-Pilson's book, *Bayesian Methods for Hackers*.

Basically, an O-ring failed during the launch, which caused the shuttle to explode.

```python
import numpy as np
import pandas as pd

challenger_data = pd.read_csv('data/challenger_data.csv')
challenger_data.dropna(inplace=True)
challenger_data.drop(labels=[24], inplace=True)
challenger_data.rename(columns={'Damage Incident': 'Damage'}, inplace=True)
challenger_data['Damage'] = challenger_data['Damage'].astype(int)
challenger_data
```

As a plot of temperture vs. damage

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(challenger_data["Temperature"], challenger_data["Damage"], 'o')
ax.xlabel="Temperture (F)"
ax.ylabel="Probability of Failure"
ax.set_title("Damage to o-Ring vs. Temperature")
```

When theoutside temperature was below 65F, there was always damage to the O-rings. Above 65, there was usually no damage.

Based on this figure, it seems plausible that the probability of damage is related to temperature. if we assume that this probability follows the logistic model, then we can write

$$
\log O(H|x) = \beta_0 + \beta_1 x
$$

Where $H$ is the hypothesis that the O-ring will be damaged, $x$ is temperature, and $\beta_0$ and $\beta_1$ are the parameters we will estimate. For reasons, define $x$ to be the temperature shifted by an offset so the mean is 0:

```python
offset = challenger_data["Temperature"].mean().round()
challenger_data['x'] = challenger_data['Temperature'] - offset
offset
```

For consistency, keep a copy of the damage column called $y$.

```python
challenger_data['y'] = challenger_data['Damage']
challenger_data.head()
```

Before doing the update, we'll use `statsmodel` to run a conventional (non-Bayesian) logistic regression:

```python
import statsmodels.formula.api as smf

formula = "y ~ x"
results = smf.logit(formula, data=challenger_data).fit(disp=False)
results.params
```

```python
# Let's compute a variety of probabilities to see what this means:

inter = results.params['Intercept']
slope = results.params['x']
xs = np.arange(53, 83) - offset

# now the log odds
log_odds = inter + xs * slope

# now to probabilities
odds = np.exp(log_odds)
ps = odds / (1 + odds)

# actually, this operation (log odds -> probabilities) is so common that it has its own scipy function
from scipy.special import expit
ps  = expit(inter + slope * xs)

# here's what the logistic model looks like with our point estimate for the parameters
fig, ax = plt.subplots()
ax.plot(challenger_data["Temperature"], challenger_data["Damage"], 'o')
ax.plot(xs + offset, ps)
ax.xlabel="Temperature (F)"
ax.ylabel="Probability of Failure"
ax.set_title("Damage to o-Ring vs. Temperature")
```

That was the conventional point estimate. Now let's do it with Bayesian methods!

## Prior Distribution

Use a uniform prior for each parameter, using the point estimates we calculated earlier to choose the upper and lower bounds:

```python
from utils import make_uniform, make_joint

qs = np.linspace(-5, 1, num=101)
prior_inter = make_uniform(qs)

qs = np.linspace(-0.8, 0.1)
prior_slope = make_uniform(qs)

# intercept across the columns, slope down the rows
joint = make_joint(prior_inter, prior_slope)

# again, conventient to `stack` the distribution:
from empiricaldist import Pmf
joint_pmf = Pmf(joint.stack())
joint_pmf.head()
```

## Likelihood

To calculate the likelihood, we'll group the data by temperature, $x$, and count the number of launches and damage incidents at each temperature.

```python
grouped = challenger_data.groupby('x')['y'].agg(['count', 'sum'])
grouped.head()
```

```python
# count is the number of launches at each temperature; sum is the number of damage incidents
# For consistency with the binomial distribution, we'll rename them ns and ks
ns = grouped['count']
ks = grouped['sum']

# to compute the likelihood, we'll assume that the parameters we calculated are correct:
xs = grouped.index
xs
```

```python
ps = expit(inter + slope * xs)
ps
```

`ps` contains the probability of damage for each  launch temperature, according to the model.

Now, for each temperature we have `ns`, `ps`, and `ks`; we can use the binomial distribution to compute the likelihood of the data.

```python
from scipy.stats import binom

likes = binom.pmf(ks, ns, ps)
likes
```

Each element of `likes` is the probability of getting `k` damage incidents in `n` launches if the probability of damage is `p`. The likelihood of the whose dataset is the product of this array:

```python
likes.prod()
```

```python
# Now let's combine all of this
likelihood = joint_pmf.copy()
for slope, inter in joint_pmf.index:
    ps = expit(inter + slope * xs)
    likes = binom.pmf(ks, ns, ps)
    likelihood[slope, inter] = likes.prod()
```

## The Update

Now we can compute the posterior distribution in the normal way:

```python
posterior_pmf = joint_pmf * likelihood
posterior_pmf.normalize()

# now to plot it
from utils import plot_contour
joint_posterior = posterior_pmf.unstack()
plot_contour(joint_posterior)
```

```python
results.params
```

```python
posterior_pmf.max_prob()
```

So we see that doing it this way is consistent with our point estimate. Because there is a slant to the oval, we can see that there is a correlation. In fact, that is one of the reasons that we centred the data - centring it minimizes the correlation. 

In theory, the correlation is not a problem, but in practice it is. With uncentred data, the results are more spread out, which means that we have less precision. By centring the data, we get better precision for the same amount of computation.

```python
from utils import marginal

marginal_inter = marginal(joint_posterior, 0)
marginal_slope = marginal(joint_posterior, 1)

marginal_inter.plot()
```

## Transforming Distributions

Let's interpret these parameters. Recall that the intercept is the log odds of the hypothesis when $x=0$, which is when `temperature` is about 70 (`offset`). So we can interpret the quantities in `marginal_inter` as log odds.

To convert them to probabilities, use the following function, which transforms the quantities in a Pmf by applying a given function:

```python
def transform(pmf, func):
    """ Apply the given transformation to the PMF """
    ps = pmf.ps
    qs = func(pmf.qs)
    return Pmf(ps, qs, copy=True)

# so now let's pass in expit to transform the log odds to probabilities
marginal_probs = transform(marginal_inter, expit)

# Of course, PMF also has a transform function that does the same thing
marginal_probs = marginal_inter.transform(expit)
marginal_probs.plot()
```

Along the bottom we have the probabilities of damage at `offset` (about 70), with the pdf along the side. This alos demonstrates another nice property of the offset: this is now the probability at a relevant temperature, rather than 0F.

Now let's look at the estimated slope. In the logistic model, the parameter $\beta_1$ is the log of the likelihood ratio. Thus, we can interpret the quantities in `marginal_slope` as log likelihood ratios, and we can use exp to transform them to likelihood ratios (Bayes factors).

```python
marginal_lr = marginal_slope.transform(np.exp)
marginal_lr.plot()
```

The $x$ axis is the likelihood ratio of 1 degree F, while the $y$ axis is the PDF. The mean is about 0.75, which means that each additional degree Fahrenheit provides evidence *against* the likelihood of damage, with a likelihood ratio (Bayes factor) of about 0.75.

Notice:
- Calculated the posterior mean of the probability of damage at 70F by first transforming the distribution, then taking the mean
- Calculate the posterior mean of the likelihood ratio by transforming the distribution, then taking the mean

This is the correct order, as opposed to taking the mean and then transforming that.

## Predictive Distributions
In the logistic model, the parameters are interpretable, at least after transformation. However, we usually care about the predictive power of the model. In this problem, the most important question we want answered is "What is the probability of O-ring damage at 31F?".

To make that prediction, let's draw a sampple of parameter pairs from the posterior distribution:

```python
sample = posterior_pmf.choice(101)
```

The result is an array of 101 tuples, each representing a possible pair of parameters. To generate the prediction, use a range of temperatures from 31F (the temperature at which Challenger launched) to 82F (the highest observed launch temperature).

```python
temps = np.arange(31, 83)
xs = temps - offset

pred = np.empty((len(sample), len(xs)))

for i, (slope, inter) in enumerate(sample):
    pred[i] = expit(inter + slope * xs)
pred.shape
```

The result has one column for each value of xs and one row for each element of samples.

In each column, calculate the median to quantify the central tendency, and a 90% credible interval to quantify the uncertainty.

```python
low, median, high = np.percentile(pred, [5, 50, 95], axis=0)
```

```python
plt.fill_between(temps, low, high, color="C1", alpha=0.2)
plt.plot(temps, median, color="C1")
plt.plot(challenger_data["Temperature"], challenger_data['Damage'], "o")
```

According to these data, the probability of damage at 80F is about 2%, but there is some uncertainty - the upper bound is 10%.

At 60F, the probability of damage is near 80%, but the CI is even wider - from 48% to 97%.

But the primary goal is to predict the probability of damage at 31F, and here we get the answer of at least 97% and most likely 99.9%.

The lesson we might learn from this is that if the engineers had considered all of the data, they might have predicted that the probability of damage was very high and tried to postpone the launch.

At the same time, if they considered the previous figure, they might have noticed that the prediction is venturing far outside the data! While our assumption that each degree F contributes the same amount of evidence for / against the probability of damage probably holds within the fairly narrow range of the data, it becomes increasingly untenable the farther outside of that we venture. As such, they might reasonably have had some doubts about this result!

## Empirical Bayes

In this chapter, we used `statsmodel` to compute the parameters that maximize the probability of the data, and then used those estimate to choose the bounds of our priors. It might occur to you that this process uses the data twice - once to choose the priors, and then again to perform the update. If that bothers you, you wouldn't be alone. The process that we used is called the Empirical Bayes method (not a great name for it).

Although it might seem problematic to use the data twice, in this case it isn't. To see why, consider the alternative. Instead of using the estimated paraters to choose the bounds, we could have used a mich wider range. In that case, the results would have been the same, we just would have spent more time computing the probabilities for parmeters here the posterior probabilities are neglibibly small.

You can think of this method as an optimization that minimizes our time computing very small probabilities.

>
> One way to answer that is to consider two hypothese:
> - `H`: the pundit's algorithm is legitimate - the probabilities it produces are correct in the sense that they accurately reflect the candidate's probability of winning
> - `not H` - the pundit's algorithm is bogus - the probabilities it produces are random values with a mean of 50%.
>
> If the pundit says Alice has a 70% chance of winning and she does, that provides evidence in favour of `H` with likelihood ration 70 / 50
>
> If the pundit says Alice has a 70% chance of winning and she loses, that provides evidence against `H` with likelihood ration 30 / 50
>
> Suppose we start with some confidence in the algorithm, so the prior odds are 4 : 1, and the pundit generates predictions for three elections:
> - First: Alice has a 70% chance of winning, and she does
> - Second: Bob has a 30% chance of winning and he does
> - Third: Carol has a 90% chance of winning and she does.
>
> What is the log-likelihood ratio for each of these outcomes? Use the log-odds form of Bayes' rule to compute the posterior log odds for `H` after each outcome. In total, do these outcomes increase or decrease your confidence in the pundit?
>
> Read more about this topic at http://allendowney.blogspot.com/2016/11/why-are-we-so-surprised.html

log(3 / 5)
log(9 / 5)

The authors of the study suggest that the cause is "age-based variation in behaviour that may be attributed to ADHD rather than the younger age of the children."
>
> Use the methods in this chapter to estimate the probability of diagnosis as a function of birth month. The notebook for this chapter provides the data and some suggestions for getting started.

### Solution 16-3


We can grab the data from the table:

```python
# number of children born in each month (starting with January)
n = np.array([32690, 31238, 34405, 34565, 34977, 34415, 
                   36577, 36319, 35353, 34405, 31285, 31617])

# number of children diagnosed with ADHD (starting with January)
k = np.array([265, 280, 307, 312, 317, 287, 
                      320, 309, 225, 240, 232, 243])

# "roll" the data so that it starts in September
x = np.arange(12)
n = np.roll(n, -8)
k = np.roll(k, -8)

# now put it in a DataFrame with one row per month and the diagnosis rate per 10 000
adhd = pd.DataFrame(dict(x=x, k=k, n=n))
adhd['rate'] = adhd['k'] / adhd['n'] * 10000
adhd
```

```python
def plot_adhd(adhd):
    fig, ax = plt.subplots()
    ax.plot(adhd['x'], adhd['rate'], 'o', 
             label='data', color='C0', alpha=0.4)
    
    plt.axvline(5.5, color='gray', alpha=0.2)
    #plt.text(6, 64, 'Younger than average')
    #plt.text(5, 64, 'Older than average', horizontalalignment='right')

    ax.set_xlabel("Birth date, months after cutoff")
    ax.set_ylabel("Diagnosis rate per 10 000")
    fig.suptitle("Title")
    
plot_adhd(adhd)
```



For the first 9 months, from September to May, we see what we would expect if some of the excess diagnoses are due to "age-based variation in behavior". For each month of difference in age, we see an increase in the number of diagnoses.

This pattern breaks down for the last three months, June, July, and August. This might be explained by random variation, but it also might be due to parental manipulation; if some parents hold back children born near the deadline, the observations for these month would include a mixture of children who are relatively old for their grade and therefore less likely to be diagnosed.

Unfortunately, the dataset includes only month of birth, not year, so we don't know the actual ages of these students when they started school. However, we can use the first nine months to estimate the effect of age on diagnosis rate; then we can think about what to do with the other three months.

Use the methods in this chapter to estimate the probability of diagnosis as a function of birth month. Start with the following prior distributions.

```python
qs = np.linspace(-5.2, -4.6, num=51)
prior_inter = make_uniform(qs, 'Intercept')

qs = np.linspace(0.0, 0.08, num=51)
prior_slope = make_uniform(qs, 'Slope')
```

1. Make a joint prior distribution and update it using the data for the first nine months.
1. Then draw a sample from the posterior distribution and use it to compute the median probability of diagnosis for each month and a 90% credible interval.
1. As a bonus exercise, do a second update using the data from the last three months, but treating the observed number of diagnoses as a lower bound on the number of diagnoses there would be if no children were kept back.

```python
# make the joint distribution
joint_prior = make_joint(prior_inter, prior_slope)
joint_prior.head()
```

```python
joint_pmf = Pmf(joint_prior.stack())

num_legit = 9
adhd1 = adhd.loc[:num_legit]
adhd2 = adhd.loc[num_legit + 1:]
adhd1
```

```python
adhd2
```

```python
from scipy.stats import binom

likelihood1 = joint_pmf.copy()

xs = adhd1['x']
ks = adhd1['k']
ns = adhd1['n']

for slope, inter in joint_pmf.index:
    ps = expit(inter + slope * xs)
    likes = binom.pmf(ks, ns, ps)
    likelihood1[slope, inter] = likes.prod()
    
likelihood1.sum()
```

```python
# This solution uses the binomial survival function to calculate the probability that the number exceeds ks
likelihood2 = joint_pmf.copy()

xs = adhd2['x']
ns = adhd2['n']
ks = adhd2['k']

for slope, inter in joint_pmf.index:
    ps = expit(inter + slope * xs)
    likes = binom.sf(ks, ns, ps)
    likelihood2[slope, inter] = likes.prod()
    
likelihood2.sum()
```

```python
posterior_pmf = joint_pmf * likelihood1
posterior_pmf.normalize()
```

```python
posterior_pmf.max_prob()
```

```python
posterior_pmf = joint_pmf * likelihood1 * likelihood2
posterior_pmf.normalize()
```

```python
posterior_pmf.max_prob()
```

```python
joint_posterior = posterior_pmf.unstack()

plot_contour(joint_posterior)
```

```python
marginal_inter = marginal(joint_posterior, 0)
marginal_slope = marginal(joint_posterior, 1)
```
