## The World Cup Problem
2018 FIFA World Cup final: France defeated Croatia 4 to 2.
1. How confident should we be that France is the better team?
2. If the same teams played again, what is the chance that France would win again?

Let's make some modelling decisions:
- Assume that for one team playing against another, there is some fixed goal-scoring rate $\lambda$
- Assume that scoring a goal is equally likely during any minute of the game
- Assume that a team never scores twice in the same minute

## Poisson Distribution
If the number of goals scored in the game follows a Poisson distribution, then the probability of scoring $k$ goals is
$$
P(k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

```python
from scipy.stats import poisson

lam = 1.4
dist = poisson(lam)

k = 4
dist.pmf(k)
```

So if the average goal-scoring rate is 1.4 / game, then the probability of scoring 4 goals wouls be about 4%.

```python
from empiricaldist import Pmf

def make_poisson_pmf(lam, qs):
    ps = poisson(lam).pmf(qs)
    pmf = Pmf(ps, qs)
    pmf.normalize()
    return pmf
```

```python
import numpy as np

lam = 1.4
goals = np.arange(10)
pmf_goals = make_poisson_pmf(lam, goals)
pmf_goals.plot(kind="bar")
```

## The Gamma Distribution

Using data from previous World Cups, estimate is that each team scores about 1.4 goals / game, on average.

Choose the Gamma distribution, because reasons.

```python
from scipy.stats import gamma

alpha = 1.4
qs = np.linspace(0, 10, 101)
ps = gamma(alpha).pdf(qs)

prior = Pmf(ps, qs)
prior.normalize()
prior.plot()
```

```python
def update_poisson(pmf, data):
    """ Update Pmf with a Poisson likelihood """
    k = data
    lams = pmf.qs
    likelihood = poisson(lams).pmf(k)
    pmf *= likelihood
    pmf.normalize()
    
france = prior.copy()
update_poisson(france, 4)
prior.plot()
france.plot()
```

So as expected, the number of goals makes larger values of lambda more likely.

```python
croatia = prior.copy()
update_poisson(croatia, 2)
prior.plot()
croatia.plot()
```

```python
print(france.mean(), croatia.mean())
```

## Probability of Superiority

Now that we have the PMFs, we can start to answer the question: what is the probability that one team is 'better' than the other, where here 'better' means 'has a larger $\lambda$'. One way: enumerate all pairs of values from the two distributions, adding up the total probability that one value exceeds the other:

```python
def prob_gt(pmf1, pmf2):
    total = 0
    for q1, p1 in pmf1.items():
        for q2, p2 in pmf2.items():
            if q1 > q2:
                total += p1 * p2
    return total

prob_gt(france, croatia)
```

```python
Pmf.prob_gt(france, croatia)
```

## Predicting the Rematch

If the teams played again, what is the probability that France would win? We'll generate the "posterior predictive distribution", which is the number of goals that we expect the teams to score.

If we knew the goal-scoring rate $\lambda$, the distribution of goals would be Poisson with that parameter. However, it is actually a mixture of different distributions with each of these values of $\lambda$.

```python
import pandas as pd

# from "General Mixtures"
def make_mixture(pmf, pmf_seq):
    """ Make a mixture of the distributions """
    df = pd.DataFrame(pmf_seq).fillna(0).transpose()
    df *= np.array(pmf)
    total = df.sum(axis=1)
    return Pmf(total)

# here we generate the sequence of PMFs
pmf_seq = [ make_poisson_pmf(lam, goals) for lam in prior.qs ] 

# The predictive distribution is a mixture of these PMF objects weighted with the posterior probabilities
pred_france = make_mixture(france, pmf_seq)
pred_france.plot(kind="bar")
```

The distribution represents two sources of uncertainty - we don't know the actual value of $\lambda$, and even if we did we would not know the number of goals in the next game.

```python
pred_croatia = make_mixture(croatia, pmf_seq)

win = Pmf.prob_gt(pred_france, pred_croatia)
print(f"Probability that France wins: {win}")

lose = Pmf.prob_lt(pred_france, pred_croatia)
print(f"Probability that France loses: {lose}")

tie = Pmf.prob_eq(pred_france, pred_croatia)
print(f"Probability that they tie: {tie}")

print(f"Assuming that France wins about half of the ties: {win + tie / 2}")
```

## The Exponential Distribution

If a process is Poisson, then the interval between events (goals) follows an exponential distribution. If the goal-scoring rate is $\lambda$, the probability of seeing an interval $t$ between goals is
$$
P(t) \alpha \lambda e ^ {-\lambda t}
$$

(Actually it just means that the probability is proportional, since the exponential is a continuous distribution. But for us we can normalize it, so we won't worry about the difference).

```python
def expo_pdf(t, lam):
    """ Compute the PDF of the exponential distribution """
    return lam * np.exp(-lam * t)

qs = np.linspace(0, 4, 101)
ps = expo_pdf(qs, lam)
pmf_time = Pmf(ps, qs)
pmf_time.normalize()

pmf_time.plot()
```

> Recommended steps:

> 2. Compute the posterior distribution of lam after the first goal.
> 3. Compute the probability of scoring another goal after another 12 minutes and do another update. Plot the prior, posterior after one goal, and posterior after both goals.
> 4. Compute the posterior predictive distribution of goals Germany might score during the remaining time in the game, 90 - 23 minutes. Note: you will have to think about how to generate predicted goals for a fraction of a game.

### Solution - Attempt 2

```python
# First we start with a gamma prior with mean 1.4
mean = 1.4
possible_lambdas = np.linspace(0, 10, 101)
prior_for_lambdas = Pmf(gamma(mean).pdf(possible_lambdas), possible_lambdas)
prior_for_lambdas.normalize()

def update_with_exponential(pmf, data):
    """ Given the pmf, update it with an exponential distribution based on the data """
    t = data / 90
    likelihood = expo_pdf(t, pmf.qs)
    #likelihood = Pmf(expo_pdf(t, pmf.qs), pmf.qs)
    pmf *= likelihood
    pmf.normalize()

germany_lambda_after_one = prior_for_lambdas.copy()
update_with_exponential(germany_lambda_after_one, 11)
germany_lambda_after_two = germany_lambda_after_one.copy()
update_with_exponential(germany_lambda_after_two, 12)
prior_for_lambdas.plot()
germany_lambda_after_one.plot()
germany_lambda_after_two.plot()
print(germany_lambda_after_one.mean(), germany_lambda_after_two.mean())
```

```python
## OK, now we have the probability distribution for lambda. 
# If we want to know the distribution for the number of goals, we now need to make a mixture of all of the Poisson
# distributions, one for each lambda, weighted by the probability of that one being the 'correct' value
possible_goals = np.arange(10)
pmf_seq = [ make_poisson_pmf(lam, possible_goals) for lam in germany_lambda_after_two.qs ]
predicted_goal_distribution = make_mixture(germany_lambda_after_two, pmf_seq)
predicted_goal_distribution.plot(kind="bar")
```

```python
## Now we want to sort out the prediction for the remaining portion of the game
# Lambda is goals / game, so we can adjust our lambda distribution proportionally and use that to
# create the mixture
fraction_remaining = (90 - 23) / 90
pmf_seq = [ make_poisson_pmf(fraction_remaining * lam, possible_goals) for lam in germany_lambda_after_two.qs]
predicted_goal_distribution_for_remaining = make_mixture(germany_lambda_after_two, pmf_seq)
predicted_goal_distribution_for_remaining.plot(kind="bar")
```

```python
## probability of scoring 5 or more goals:
predicted_goal_distribution_for_remaining.prob_ge(5)
```

### Solution

```python
qs = np.linspace(0, 10, 101)
ps = gamma(1.4).pdf(qs)
prior = Pmf(ps, qs)
prior.normalize()

def update_exponential(pmf, data):
    """ Update a PMF with an exponential likelihood """
    t = data
    qs = np.linspace(0, 10, 101)
    ps = expo_pdf(t, qs)
    likelihood = Pmf(ps, qs)
    pmf *= likelihood
    pmf.normalize()
    
after_one = prior.copy()
update_exponential(after_one, 11 / 90)

after_two = after_one.copy()
update_exponential(after_two, 12 / 90)

prior.plot()
after_one.plot()
after_two.plot()

print(prior.mean(), after_one.mean(), after_two.mean())
```

```python
time_remaining = (90 -23) / 90
pmf_seq = [ make_poisson_pmf(lam * time_remaining, goals) for lam in after_two.qs ]
predictive = make_mixture(after_two, pmf_seq)
predictive.plot(kind="bar")
```

### 8.2
> Return to the France-Croatia World Cup, and suppose that they play a rematch. What is the probability that France scores first?

```python
def make_exponential_pmf(lam, high):
    qs = np.linspace(0, high, 101)
    ps = expo_pdf(qs, lam)
    pmf = Pmf(ps, qs)
    pmf.normalize()
    return pmf

# Why are we choosing to set the upper limit at 4? Because this the upper limit for t (time between goals,
# measured in games - seems unlikely that a team will go for four games without a goal. Changing this doesn't
# change the actual answer too much.)
pmf_seq = [ make_exponential_pmf(lam, 4) for lam in prior.qs ]

pred_france = make_mixture(france, pmf_seq)
pred_croatia = make_mixture(croatia, pmf_seq)

Pmf.prob_lt(pred_france, pred_croatia)
```

### 8.3
> In the 2010 - 11 NHL Finals, Bruins against Canucks. Boston lost the first two games 0 - 1 and 2 - 3, then won the next two games 8 - 1 and 4 - 0. At this point, what is the probability that the Bruins will win the next game, and what is the probability of winning the championship?
>
> To choose the prior distribution, using statistics from nhl.com, distribution of goals is well-modelled by a Gamma distribution with mean of 2.8.
>
> In what ways do you think that the outcome of these games might violate the assumptions of the Poisson model? How would these violations affect your predictions?

### Solution

```python
from scipy.stats import gamma

alpha = 2.8
qs = np.linspace(0, 15, 101)
ps = gamma.pdf(qs, alpha)
prior_hockey = Pmf(ps, qs)
prior_hockey.normalize()
prior_hockey.plot()
print(prior_hockey.mean())
```

```python
bruins = prior_hockey.copy()
for data in [0, 2, 8, 4]:
    update_poisson(bruins, data)
bruins.mean()
```

```python
canucks = prior_hockey.copy()
for data in [1, 3, 1, 0]:
    update_poisson(canucks, data)
canucks.mean()
```

```python
bruins.plot(label="Bruins")
canucks.plot(label="Canucks")
```

```python
goals = np.arange(15)
pmf_seq = [ make_poisson_pmf(lam, goals) for lam in bruins.qs]
```

```python
pred_bruins = make_mixture(bruins, pmf_seq)
pred_canucks = make_mixture(canucks, pmf_seq)
```

```python
pred_bruins.plot(kind="bar")
```

```python
pred_canucks.plot(kind="bar")
```

```python
win = Pmf.prob_gt(pred_bruins, pred_canucks)
lose = Pmf.prob_lt(pred_bruins, pred_canucks)
tie = Pmf.prob_eq(pred_bruins, pred_canucks)

prob_win = win + tie / 2
prob_win
```
